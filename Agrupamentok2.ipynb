```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score
from openpyxl import Workbook

# Ler o arquivo CSV
data = pd.read_csv('Escala_k2_ns.csv' , encoding='latin-1')

# Separar os clusters
cluster_0 = data[data['Cluster'] == 'cluster0']
cluster_1 = data[data['Cluster'] == 'cluster1']

plt.scatter(cluster_0['Instance_number'], cluster_0['Gênero=Mulher'], color='red', label='Cluster 0')
plt.scatter(cluster_1['Instance_number'], cluster_1['Gênero=Mulher'], color='blue', label='Cluster 1')
plt.xlabel('Instance_number')
plt.ylabel('Gênero=Mulher')
plt.title('Clusters')
plt.legend()
plt.show()

# Comparar os clusters
silhouette_score_cluster_0 = silhouette_score(cluster_0[['Instance_number', 'Gênero=Mulher']], cluster_0['cluster'])
silhouette_score_cluster_1 = silhouette_score(cluster_1[['Instance_number', 'Gênero=Mulher']], cluster_1['cluster'])

print(f"Silhouette Score para o Cluster 0: {silhouette_score_cluster_0}")
print(f"Silhouette Score para o Cluster 1: {silhouette_score_cluster_1}")
```

```python
# Escolha do atributo para avaliação
atributo_a_avaliar = 'Gênero=Mulher'  # Substitua 'atributo1' pelo atributo que deseja avaliar

# Contar o número de ocorrências do atributo em cada cluster
quantidade_cluster_0 = cluster_0[atributo_a_avaliar].count()
quantidade_cluster_1 = cluster_1[atributo_a_avaliar].count()

print(f'Quantidade de "{atributo_a_avaliar}" no Cluster 0: {quantidade_cluster_0}')
print(f'Quantidade de "{atributo_a_avaliar}" no Cluster 1: {quantidade_cluster_1}')

# Determinar qual cluster possui o atributo em maior quantidade
if quantidade_cluster_0 > quantidade_cluster_1:
    print(f'O Cluster 0 possui o atributo "{atributo_a_avaliar}" em maior quantidade.')
elif quantidade_cluster_1 > quantidade_cluster_0:
    print(f'O Cluster 1 possui o atributo "{atributo_a_avaliar}" em maior quantidade.')
else:
    print(f'A quantidade do atributo "{atributo_a_avaliar}" é igual nos dois clusters.')
```

```python
# Calcular a média do atributo em cada cluster
media_cluster_0 = cluster_0[atributo_a_avaliar].mean()
media_cluster_1 = cluster_1[atributo_a_avaliar].mean()

# Mostrar os valores das médias
print(f'Média de "{atributo_a_avaliar}" no Cluster 0: {media_cluster_0:.2f}')
print(f'Média de "{atributo_a_avaliar}" no Cluster 1: {media_cluster_1:.2f}')

# Determinar qual cluster possui a média do atributo maior
if media_cluster_0 > media_cluster_1:
    print(f'O Cluster 0 possui uma média maior do atributo "{atributo_a_avaliar}".')
elif media_cluster_1 > media_cluster_0:
    print(f'O Cluster 1 possui uma média maior do atributo "{atributo_a_avaliar}".')
else:
    print(f'A média do atributo "{atributo_a_avaliar}" é igual nos dois clusters.')
```

```python
atributos_a_avaliar = ['Curso=Sistemas de Informação','Gênero=Homem','Gênero=Queer','Gênero=Mulher','Sexualidade=Prefiro não responder','Sexualidade=Homossexual','Sexualidade=Heterossexual','Sexualidade=Pansexual','Sexualidade=Bissexual','Sexualidade=Assexual','Assédio na FACOM=Sim','VG2','VM3','VG3','VG4','VG5','VM1','VG1','VG6','IV1','IV2','VM2']
wb = Workbook()
ws = wb.active
ws.append(['Atributo', 'Média Cluster 0', 'Média Cluster 1'])


for atributo in atributos_a_avaliar:
    # Calcular a média do atributo em cada cluster
    media_cluster_0 = cluster_0[atributo].mean()
    media_cluster_1 = cluster_1[atributo].mean()

    # Mostrar os valores das médias
    print(f'Média de "{atributo}" no Cluster 0: {media_cluster_0:.2f}')
    print(f'Média de "{atributo}" no Cluster 1: {media_cluster_1:.2f}')

    # Determinar qual cluster possui a média do atributo maior
    if media_cluster_0 > media_cluster_1:
        print(f'O Cluster 0 possui uma média maior do atributo "{atributo}".')
    elif media_cluster_1 > media_cluster_0:
        print(f'O Cluster 1 possui uma média maior do atributo "{atributo}".')
    else:
        print(f'A média do atributo "{atributo}" é igual nos dois clusters.')
    print()  # Adiciona uma linha em branco entre os resultados de diferentes atributos
    ws.append([atributo, media_cluster_0, media_cluster_1])

wb.save('medias_atributos_k2.xlsx')
```
